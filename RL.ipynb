```python
# ç®€ç‰ˆ
import numpy as np
import pandas as pd

class RewardGenerator:
    def __init__(self, volatility_target, transaction_cost_bp, mu=1.0):
        """
        åˆå§‹åŒ–å¥–åŠ±è®¡ç®—å™¨
        
        å‚æ•°:
            volatility_target (float): ç›®æ ‡æ³¢åŠ¨ç‡ (sigma_tgt)
            transaction_cost_bp (float): äº¤æ˜“æˆæœ¬åŸºç‚¹ (ä¾‹å¦‚ 10bp = 0.0010)
            mu (float): å½’ä¸€åŒ–å¸¸æ•°ï¼Œè®ºæ–‡é»˜è®¤ä¸º 1
        """
        self.vol_tgt = volatility_target
        self.bp = transaction_cost_bp
        self.mu = mu
        
        # ç¼“å­˜ä¸Šä¸€æ­¥å’Œä¸Šä¸Šæ­¥çš„çŠ¶æ€
        self.prev_position_scaled = 0.0  # å¯¹åº” (sigma_tgt / sigma_t-2) * A_t-2
        self.prev_price = None           # p_t-1
    
    def reset(self, initial_price):
        """åœ¨æ¯ä¸ª episode å¼€å§‹æ—¶é‡ç½®"""
        self.prev_position_scaled = 0.0
        self.prev_price = initial_price

    def calculate_step_reward(self, current_price, current_vol_est, action_prev):
        """
        è®¡ç®—å•æ­¥å¥–åŠ± R_t
        
        å‚æ•°:
            current_price (float): å½“å‰æ—¶åˆ» t çš„ä»·æ ¼ p_t
            current_vol_est (float): t-1 æ—¶åˆ»ä¼°ç®—çš„æ³¢åŠ¨ç‡ sigma_t-1
                                     (æ³¨æ„ï¼šè¿™æ˜¯ä½ åœ¨åšå†³å®š A_t-1 æ—¶èƒ½çœ‹åˆ°çš„æ³¢åŠ¨ç‡)
            action_prev (float): t-1 æ—¶åˆ»åšå‡ºçš„åŠ¨ä½œ A_t-1 (èŒƒå›´ -1 åˆ° 1)
        
        è¿”å›:
            reward (float): R_t
        """
        if self.prev_price is None:
            raise ValueError("è¯·å…ˆè°ƒç”¨ reset() åˆå§‹åŒ–ä»·æ ¼")

        # 1. è®¡ç®—åŠ æ³•æ”¶ç›Š r_t = p_t - p_{t-1}
        # è®ºæ–‡æ˜ç¡®æŒ‡å‡ºä½¿ç”¨ additive profits
        r_t = current_price - self.prev_price
        
        # 2. è®¡ç®—æ³¢åŠ¨ç‡ç¼©æ”¾ç³»æ•° (Scaling Factor)
        # scale = sigma_tgt / sigma_{t-1}
        # é¿å…é™¤ä»¥é›¶
        vol_scale = self.vol_tgt / (current_vol_est + 1e-8)
        
        # 3. è®¡ç®—å½“å‰çš„ç¼©æ”¾åä»“ä½ (Scaled Position)
        # current_scaled_pos = (sigma_tgt / sigma_{t-1}) * A_{t-1}
        current_position_scaled = vol_scale * action_prev
        
        # 4. è®¡ç®—æ¯›åˆ©æ¶¦ (Gross Profit)
        # Term 1: A_{t-1} * scale * r_t
        gross_profit = current_position_scaled * r_t
        
        # 5. è®¡ç®—äº¤æ˜“æˆæœ¬ (Transaction Cost)
        # Cost = bp * p_{t-1} * | current_scaled_pos - prev_scaled_pos |
        # æ³¨æ„ï¼šè¿™é‡Œè®¡ç®—çš„æ˜¯æŒä»“è§„æ¨¡çš„å˜åŒ–é‡ (Turnover)
        turnover = np.abs(current_position_scaled - self.prev_position_scaled)
        cost = self.bp * self.prev_price * turnover
        
        # 6. è®¡ç®—æœ€ç»ˆå¥–åŠ± R_t
        reward = self.mu * (gross_profit - cost)
        
        # --- æ›´æ–°çŠ¶æ€ç”¨äºä¸‹ä¸€æ­¥ ---
        self.prev_price = current_price
        self.prev_position_scaled = current_position_scaled
        
        return reward

# ==========================================
# ä½¿ç”¨ç¤ºä¾‹ (æ¨¡æ‹Ÿ Gym ç¯å¢ƒçš„ä¸€ä¸ª Step)
# ==========================================

# 1. é…ç½®å‚æ•°
# å‡è®¾ç›®æ ‡æ³¢åŠ¨ç‡æ˜¯æ¯å¤© 2.0 (å¦‚æœæ˜¯ä»·æ ¼å˜åŠ¨çš„stdï¼Œå•ä½æ˜¯è´§å¸; å¦‚æœæ˜¯æ”¶ç›Šç‡stdï¼Œéœ€æ¢ç®—)
# è®ºæ–‡ä¸­ r_t æ˜¯ä»·æ ¼å˜åŠ¨ï¼Œæ‰€ä»¥ sigma ä¹Ÿåº”è¯¥æ˜¯ä»·æ ¼å˜åŠ¨çš„ std
VOL_TARGET = 5.0  
COST_BP = 0.0001  # 1bp
reward_gen = RewardGenerator(volatility_target=VOL_TARGET, transaction_cost_bp=COST_BP)

# æ¨¡æ‹Ÿæ•°æ®æµ
prices = [100.0, 102.0, 101.5, 103.0]
# å‡è®¾è¿™æ˜¯é€šè¿‡ rolling(60).std() è®¡ç®—å‡ºçš„ä»·æ ¼å˜åŠ¨æ ‡å‡†å·®
vols = [1.5, 1.8, 1.6, 1.7] 
actions = [1, 0.5, -0.5] # å¯¹åº” t=0, t=1, t=2 çš„åŠ¨ä½œ

# åˆå§‹åŒ–
reward_gen.reset(initial_price=prices[0])

print(f"Initial Price: {prices[0]}")

# æ—¶é—´æ­¥å¾ªç¯
for t in range(1, len(prices)):
    p_t = prices[t]
    
    # è·å– t-1 æ—¶åˆ»çš„ä¿¡æ¯ (è¿™æ˜¯æˆ‘ä»¬åœ¨ t æ—¶åˆ»è®¡ç®— R_t æ‰€ä¾æ®çš„è¿‡å»å†³ç­–ä¿¡æ¯)
    sigma_t_minus_1 = vols[t-1]
    action_t_minus_1 = actions[t-1]
    
    # è®¡ç®—å¥–åŠ±
    r_t = reward_gen.calculate_step_reward(
        current_price=p_t,
        current_vol_est=sigma_t_minus_1,
        action_prev=action_t_minus_1
    )
    
    print(f"Step {t}: Price {p_t}, Action(t-1) {action_t_minus_1}, Reward: {r_t:.4f}")
```

```python
#å®ç›˜é€»è¾‘

import gym
from gym import spaces
import pandas as pd
import numpy as np
import random

# --- è®ºæ–‡è¶…å‚æ•° (å‚è€ƒè®ºæ–‡ Table 1 å’Œ Section 4) ---
# ç›®æ ‡æ³¢åŠ¨ç‡ (Sigma Target)ã€‚æ³¨æ„ï¼šå¦‚æœä½ çš„æ•°æ®æ˜¯åˆ†é’Ÿçº¿ï¼Œè¿™ä¸ªå€¼éœ€è¦è°ƒå°ï¼›
# å¦‚æœæ˜¯æ—¥çº¿ï¼Œè®ºæ–‡å¸¸ç”¨ 5.0 (ä»£è¡¨5ä¸ªç‚¹æ•°çš„æ³¢åŠ¨) æˆ–æ ¹æ®å…·ä½“åˆçº¦è°ƒæ•´ã€‚
VOLATILITY_TARGET = 5.0 
# äº¤æ˜“æˆæœ¬ (Basis Points)ã€‚è®ºæ–‡ä¸­è®­ç»ƒç”¨ 20bp (0.0020)ï¼Œæµ‹è¯•åˆ†æäº† 1bp-45bpã€‚
TRANSACTION_BP = 0.0020  
# å½’ä¸€åŒ–å¸¸æ•°ï¼Œè®ºæ–‡è®¾ä¸º 1
MU = 1.0
# æ³¢åŠ¨ç‡è®¡ç®—çª—å£ã€‚è®ºæ–‡ç”¨ 60-dayã€‚å¦‚æœæ˜¯åˆ†é’Ÿçº¿ï¼Œè¯·è°ƒæ•´è¿™ä¸ªçª—å£å¤§å° (ä¾‹å¦‚ 60*24*4)
VOL_WINDOW = 60 
MAX_LEVERAGE = 10.0   # æˆ– 5.0ï¼Œå–å†³äºä½ åˆçº¦çš„ä¿è¯é‡‘é€»è¾‘

class FutureTradingEnv(gym.Env):
    """
    åŸºäºè®ºæ–‡ 'Deep Reinforcement Learning for Trading' é€»è¾‘ä¿®æ”¹çš„ç¯å¢ƒ
    ç‰¹ç‚¹ï¼šæ— èµ„é‡‘é™åˆ¶ã€æ³¢åŠ¨ç‡ç¼©æ”¾ä»“ä½ã€åŠ æ³•åˆ©æ¶¦å¥–åŠ±
    """
    metadata = {'render.modes': ['human']}

    def __init__(self, df):
        super(FutureTradingEnv, self).__init__()

        # 1. æ•°æ®é¢„å¤„ç†
        self.df = df.copy().reset_index(drop=True)
        self._calculate_volatility() # é¢„è®¡ç®—æ³¢åŠ¨ç‡

        # 2. åŠ¨ä½œç©ºé—´
        # è®ºæ–‡ A2C è¾“å‡ºè¿ç»­å€¼ [-1, 1]ï¼Œä»£è¡¨åšå¤š/åšç©ºçš„æ–¹å‘å’Œä¿¡å¿ƒ
        self.action_space = spaces.Box(
            low=np.array([-1]), high=np.array([1]), dtype=np.float32)

        # 3. è§‚æµ‹ç©ºé—´
        # ä¿æŒä½ çš„ (28, 120) ç»“æ„ï¼Œä½†åœ¨ _next_observation ä¸­éœ€è¦æ›¿æ¢æ‰ balance ç­‰å­—æ®µ
        self.observation_space = spaces.Box(
            low=-np.inf, high=np.inf, shape=(28, 120), dtype=np.float32)

        # æ—¶é—´ç´¢å¼•å¤„ç† (ä¿æŒä½ çš„åŸæœ‰é€»è¾‘)
        self.open_time_index = list(self.df[(self.df['date'].dt.hour == 21) & (self.df['date'].dt.minute == 1)].index)
        # è¿‡æ»¤æ‰æœ«å°¾ä¸å¤Ÿé•¿çš„æ•°æ®
        self.open_time_index = [i for i in self.open_time_index if i < len(self.df) - 500]
        self._std_mean_()

        # å†…éƒ¨çŠ¶æ€è®°å½•
        self.prev_action = 0.0          # ä¸Šä¸€æ­¥çš„åŸå§‹åŠ¨ä½œ A_{t-1}
        self.prev_scaled_position = 0.0 # ä¸Šä¸€æ­¥çš„å®é™…æŒä»“ (ç¼©æ”¾å)
        
        # åˆå§‹åŒ–æ—¶é—´ç´¢å¼•
        self.start_time = 0
        self.end_time = len(self.df) - 1
        self.current_step = 0

    def _calculate_volatility(self):
        """
        æ ¹æ®è®ºæ–‡è®¡ç®— Ex-ante Volatility (sigma_t)
        è®ºæ–‡æ–¹æ³•: 60-day EWM std of additive returns (p_t - p_{t-1})
        """
        # è®¡ç®—åŠ æ³•æ”¶ç›Š (ä»·æ ¼å˜åŒ–é‡)
        price_diff = self.df['close'].diff()
        
        # è®¡ç®—æŒ‡æ•°ç§»åŠ¨æ ‡å‡†å·®ã€‚
        # shift(1) æå…¶é‡è¦ï¼å› ä¸ºåœ¨ t æ—¶åˆ»åšå†³ç­–æ—¶ï¼Œæˆ‘ä»¬åªèƒ½çœ‹åˆ° t-1 åŠä¹‹å‰çš„æ³¢åŠ¨ç‡ã€‚
        # å¦åˆ™å°±æ˜¯æœªæ¥å‡½æ•°ã€‚
        self.df['sigma_t'] = price_diff.ewm(span=VOL_WINDOW, adjust=False).std().shift(1)
        
        # å¡«å…… NaN å’Œ å¤„ç† 0 æ³¢åŠ¨ç‡ (é˜²æ­¢é™¤ä»¥0) - ä¿®å¤é—®é¢˜2å’Œ5
        self.df['sigma_t'] = self.df['sigma_t'].bfill()  # ä½¿ç”¨æ–°æ–¹æ³•æ›¿ä»£fillna(method='bfill')
        self.df['sigma_t'] = self.df['sigma_t'].clip(lower=1e-6)  # å‘é‡åŒ–æ“ä½œæ›¿ä»£apply

    def _std_mean_(self):
        """è®¡ç®—ç»Ÿè®¡é‡ç”¨äºå½’ä¸€åŒ– (ä¿æŒåŸé€»è¾‘ä¼˜åŒ–)"""
        # æ’é™¤éæ•°å€¼åˆ—
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns
        self.mean_dir = self.df[numeric_cols].mean().to_dict()
        self.std_dir = self.df[numeric_cols].std().to_dict()
        # è¡¥å…… dict key æ ¼å¼ä»¥åŒ¹é… _next_observation
        self.mean_dir = {f'{k}_mean': v for k, v in self.mean_dir.items()}
        self.std_dir = {f'{k}_std': v for k, v in self.std_dir.items()}

    def _next_observation(self):
        """
        æå–è¿‡å»120ä¸ªæ—¶é—´æ­¥çš„ç‰¹å¾
        ä¿®å¤é—®é¢˜1ï¼šå®é™…æå–ç‰¹å¾æ•°æ®è€Œéå ä½ç¬¦
        """
        # è®¡ç®—ç´¢å¼•èŒƒå›´
        end_idx = self.current_step + self.start_time
        start_idx = max(0, end_idx - 119)  # ç¡®ä¿ä¸è¶Šç•Œ
        
        # è·å–æ•°å€¼åˆ—ï¼ˆæ’é™¤dateç­‰éæ•°å€¼åˆ—ï¼‰
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
        
        # æå–ç‰¹å¾æ•°æ®
        if start_idx <= end_idx and end_idx < len(self.df):
            obs_data = self.df.loc[start_idx:end_idx, numeric_cols].values
        else:
            # å¦‚æœç´¢å¼•è¶Šç•Œï¼Œè¿”å›é›¶çŸ©é˜µ
            obs_data = np.zeros((120, len(numeric_cols)))
        
        # ç¡®ä¿æœ‰120è¡Œæ•°æ®ï¼ˆå¦‚æœä¸è¶³åˆ™å¡«å……ï¼‰
        if obs_data.shape[0] < 120:
            padding = np.zeros((120 - obs_data.shape[0], obs_data.shape[1]))
            obs_data = np.vstack([padding, obs_data])
        else:
            obs_data = obs_data[-120:]  # åªå–æœ€å120è¡Œ
        
        # å½’ä¸€åŒ–å¤„ç†ï¼ˆä½¿ç”¨é¢„è®¡ç®—çš„å‡å€¼å’Œæ ‡å‡†å·®ï¼‰
        for i, col in enumerate(numeric_cols):
            mean_key = f'{col}_mean'
            std_key = f'{col}_std'
            if mean_key in self.mean_dir and std_key in self.std_dir:
                mean_val = self.mean_dir[mean_key]
                std_val = self.std_dir[std_key]
                if std_val > 1e-8:  # é¿å…é™¤ä»¥0
                    obs_data[:, i] = (obs_data[:, i] - mean_val) / std_val
        
        # æ·»åŠ prev_actionä½œä¸ºé¢å¤–ç‰¹å¾ï¼ˆæ”¾åœ¨æœ€åä¸€åˆ—ï¼‰
        obs_data = np.column_stack([obs_data, np.full(120, self.prev_action)])
        
        # ç¡®ä¿å½¢çŠ¶ä¸º (28, 120) - å¦‚æœç‰¹å¾æ•°ä¸è¶³28ï¼Œç”¨0å¡«å……ï¼›å¦‚æœè¶…è¿‡ï¼Œæˆªå–å‰28åˆ—
        if obs_data.shape[1] < 28:
            padding = np.zeros((120, 28 - obs_data.shape[1]))
            obs_data = np.hstack([obs_data, padding])
        else:
            obs_data = obs_data[:, :28]
        
        # è½¬ç½®ä¸º (28, 120) æ ¼å¼
        obs = obs_data.T.astype(np.float32)
        
        return obs

    def step(self, action):
        """
        æ‰§è¡Œä¸€æ­¥äº¤æ˜“
        æ ¸å¿ƒä¿®æ”¹ï¼šæ³¢åŠ¨ç‡ç¼©æ”¾ + è®ºæ–‡å¥–åŠ±å‡½æ•°
        """
        # 1. è·å–å½“å‰æ—¶åˆ» t çš„ç´¢å¼•
        current_idx = self.current_step + self.start_time
        next_idx = current_idx + 1

        # æ£€æŸ¥æ˜¯å¦ç»“æŸ
        if next_idx >= len(self.df) or self.current_step >= (self.end_time - self.start_time - 1):
            return self._next_observation(), 0, True, {}

        # 2. è·å–ä»·æ ¼å’Œæ³¢åŠ¨ç‡
        # p_{t-1}: ä¸Šä¸€æ­¥çš„ä»·æ ¼ (å†³ç­–åŸºå‡†)
        price_prev = self.df.loc[current_idx, 'close']
        # p_t: å½“å‰è¿™ä¸€æ­¥ç»“æŸæ—¶çš„ä»·æ ¼ (è®¡ç®—ç›ˆäºç”¨)
        price_curr = self.df.loc[next_idx, 'close']
        # sigma_{t-1}: å†³ç­–æ—¶å·²çŸ¥çš„æ³¢åŠ¨ç‡
        sigma_prev = self.df.loc[current_idx, 'sigma_t']

        # 3. åŠ¨ä½œå¤„ç† (Continuous Action)
        # è®ºæ–‡ A2C è¾“å‡º [-1, 1]
        action_val = np.clip(action[0], -1, 1)

        # 4. === è®ºæ–‡æ ¸å¿ƒé€»è¾‘ï¼šæ³¢åŠ¨ç‡ç¼©æ”¾ (Volatility Scaling) ===
        # å…¬å¼: Position = Action * (Target_Vol / Realized_Vol)
        scaling_factor = np.clip(
            VOLATILITY_TARGET / sigma_prev,
            0.0,
            MAX_LEVERAGE
        )

        current_scaled_position = action_val * scaling_factor

        # 5. === è®ºæ–‡æ ¸å¿ƒé€»è¾‘ï¼šå¥–åŠ±å‡½æ•°è®¡ç®— (Equation 4) ===
        
        # (a) åŠ æ³•åˆ©æ¶¦ (Additive Return)
        # r_t = p_t - p_{t-1}
        r_t = price_curr - price_prev
        
        # (b) æ¯›åˆ©æ¶¦ (Gross Profit) - ä¿®å¤é—®é¢˜3ï¼šä½¿ç”¨ä¸Šä¸€æ­¥çš„ä»“ä½è®¡ç®—åˆ©æ¶¦
        # Profit = Position_{t-1} * r_t
        # æ³¨æ„ï¼šåˆ©æ¶¦åº”è¯¥åŸºäºä¸Šä¸€æ­¥å·²æŒæœ‰çš„ä»“ä½ï¼Œè€Œéå½“å‰ç›®æ ‡ä»“ä½
        gross_profit = self.prev_scaled_position * r_t
        
        # (c) äº¤æ˜“æˆæœ¬ (Transaction Cost) - ä¿®å¤é—®é¢˜6å’Œè®ºæ–‡ä¸€è‡´æ€§ï¼šè€ƒè™‘æ–¹å‘æ”¹å˜çš„æˆæœ¬
        # Cost = bp * p_t * |Position_t - Position_{t-1}|
        # æ³¨æ„ï¼šè®ºæ–‡ä¸­ä½¿ç”¨å½“å‰ä»·æ ¼ P_t è®¡ç®—äº¤æ˜“æˆæœ¬ï¼Œè€Œé P_{t-1}
        # å¦‚æœæ”¹å˜æ–¹å‘ï¼ˆä»åšå¤šå˜åšç©ºæˆ–åä¹‹ï¼‰ï¼Œéœ€è¦å¹³ä»“+å¼€ä»“æˆæœ¬
        position_change = current_scaled_position - self.prev_scaled_position
        
        # åˆ¤æ–­æ˜¯å¦æ”¹å˜æ–¹å‘ï¼ˆç¬¦å·ç›¸åï¼‰
        if (current_scaled_position * self.prev_scaled_position < 0):
            # æ–¹å‘æ”¹å˜ï¼šå¹³ä»“æˆæœ¬ + å¼€ä»“æˆæœ¬
            # ä½¿ç”¨å½“å‰ä»·æ ¼ price_curr
            cost = TRANSACTION_BP * price_curr * (
                abs(self.prev_scaled_position) + abs(current_scaled_position)
            )
        else:
            # åŒæ–¹å‘è°ƒæ•´ï¼šåªè®¡ç®—å˜åŒ–é‡
            # ä½¿ç”¨å½“å‰ä»·æ ¼ price_curr
            cost = TRANSACTION_BP * price_curr * abs(position_change)
        
        # (d) æœ€ç»ˆå¥–åŠ± (Net Reward)
        reward = MU * (gross_profit - cost)

        # 6. æ›´æ–°çŠ¶æ€
        self.prev_action = action_val
        self.prev_scaled_position = current_scaled_position
        self.current_step += 1
        
        # 7. æ£€æŸ¥ Done
        # è®ºæ–‡é€»è¾‘ä¸­ï¼Œé™¤éæ•°æ®è·‘å®Œï¼Œå¦åˆ™ä¸ä¼šå› ä¸ºäºæŸè€Œæå‰ Done
        done = False
        
        # 8. è·å–ä¸‹ä¸€ä¸ª Observation
        obs = self._next_observation()

        return obs, reward, done, {
            'gross_profit': gross_profit,
            'cost': cost,
            'price': price_curr,
            'scaled_pos': current_scaled_position
        }

    def reset(self, new_df=None):
        # è¿™é‡Œçš„é€»è¾‘ä¸»è¦æ˜¯é‡ç½®æ¸¸æ ‡å’Œå†…éƒ¨çŠ¶æ€
        if new_df is not None:
            self.df = new_df.copy().reset_index(drop=True)
            self._calculate_volatility()
            self.open_time_index = list(self.df[(self.df['date'].dt.hour == 21) & (self.df['date'].dt.minute == 1)].index)
            self.open_time_index_len = range(len(self.open_time_index) - 1)
            self._std_mean_()

        # é‡ç½®çŠ¶æ€å˜é‡
        self.prev_action = 0.0
        self.prev_scaled_position = 0.0
        
        # éšæœºé€‰æ‹©èµ·ç‚¹ (ä¿æŒä½ çš„åŸæœ‰é€»è¾‘)
        if len(self.open_time_index) > 1:
            index_ = random.choice(range(len(self.open_time_index) - 1))
            self.start_time = self.open_time_index[index_]
            self.end_time = self.open_time_index[index_ + 1]
            
            # ç®€å•çš„æ ¡éªŒï¼Œé˜²æ­¢ç´¢å¼•è¶Šç•Œ
            if self.end_time >= len(self.df):
                self.end_time = len(self.df) - 1
        else:
            self.start_time = 0
            self.end_time = len(self.df) - 1

        self.current_step = 0
        
        return self._next_observation()

    def render(self, mode='human', close=False):
        print(f'Step: {self.current_step}, '
              f'Action: {self.prev_action:.2f}, '
              f'ScaledPos: {self.prev_scaled_position:.2f}')
```

```python
# è®ºæ–‡é€»è¾‘
def step(self, action):
    """
    æ‰§è¡Œä¸€æ­¥äº¤æ˜“
    é€»è¾‘ï¼šåœ¨ t æ—¶åˆ»åŸºäº Observation é€‰æ‹© Actionï¼ŒæŒæœ‰åˆ° t+1 æ—¶åˆ»ç»“ç®—ã€‚
    è¿”å›ï¼št+1 æ—¶åˆ»çš„çŠ¶æ€å’Œ t åˆ° t+1 æœŸé—´äº§ç”Ÿçš„ Rewardã€‚
    """
    # 1. æ—¶é—´ç´¢å¼•
    # current_idx = t
    current_idx = self.current_step + self.start_time
    # next_idx = t + 1
    next_idx = current_idx + 1

    # æ£€æŸ¥è¾¹ç•Œ
    if next_idx >= len(self.df) or self.current_step >= (self.end_time - self.start_time - 1):
        return self._next_observation(), 0, True, {}

    # 2. è·å–æ•°æ®
    # price_prev = P_t (äº¤æ˜“æ‰§è¡Œä»·æ ¼)
    price_prev = self.df.loc[current_idx, 'close']
    # price_curr = P_{t+1} (ç»“ç®—ä»·æ ¼)
    price_curr = self.df.loc[next_idx, 'close']
    # sigma_prev = sigma_t (å†³ç­–æ—¶å·²çŸ¥çš„æ³¢åŠ¨ç‡)
    sigma_prev = self.df.loc[current_idx, 'sigma_t']

    # 3. åŠ¨ä½œå¤„ç† (Action_t)
    action_val = np.clip(action[0], -1, 1)

    # 4. æ³¢åŠ¨ç‡ç¼©æ”¾ (Volatility Scaling)
    # å¯¹åº”è®ºæ–‡ä¸­çš„ (sigma_tgt / sigma_{t-1} * A_{t-1}) åœ¨ t æ—¶åˆ»çš„ç‰ˆæœ¬
    # å³: Position_t = Action_t * (Target / sigma_t)
    scaling_factor = np.clip(
        VOLATILITY_TARGET / sigma_prev,
        0.0,
        MAX_LEVERAGE
    )
    current_scaled_position = action_val * scaling_factor

    # 5. å¥–åŠ±å‡½æ•°è®¡ç®— (è®¡ç®— R_{t+1})
    
    # (a) ä»·æ ¼å˜åŒ– r_{t+1} = P_{t+1} - P_t
    r_t = price_curr - price_prev
    
    # (b) æ¯›åˆ©æ¶¦ (Gross Profit)
    # !!! ä¿®æ­£ç‚¹1ï¼šåˆ©æ¶¦ç”±å½“å‰åŠ¨ä½œ current_scaled_position å†³å®š !!!
    # é€»è¾‘ï¼šæˆ‘ä»¬åœ¨ t æ—¶åˆ»å»ºç«‹ä»“ä½ï¼ŒæŒæœ‰åˆ° t+1ï¼Œäº«å—è¿™æ®µæ—¶é—´çš„ä»·æ ¼å˜åŠ¨ã€‚
    gross_profit = current_scaled_position * r_t
    
    # (c) äº¤æ˜“æˆæœ¬ (Transaction Cost)
    # !!! ä¿®æ­£ç‚¹2ï¼šæˆæœ¬åŸºäºäº¤æ˜“å‘ç”Ÿçš„æ—¶åˆ» t çš„ä»·æ ¼ (price_prev) !!!
    # !!! ä¿®æ­£ç‚¹3ï¼šç›´æ¥ä½¿ç”¨ç»å¯¹å€¼å·®è®¡ç®—æ¢æ‰‹ç‡ï¼Œæ— éœ€ if/else !!!
    # Change = |Position_t - Position_{t-1}|
    position_change = abs(current_scaled_position - self.prev_scaled_position)
    cost = TRANSACTION_BP * price_prev * position_change
    
    # (d) æœ€ç»ˆå¥–åŠ± (Net Reward)
    reward = MU * (gross_profit - cost)

    # 6. æ›´æ–°çŠ¶æ€
    self.prev_action = action_val
    self.prev_scaled_position = current_scaled_position
    self.current_step += 1
    
    # 7. æ£€æŸ¥ Done (é€šå¸¸åªåœ¨æ•°æ®ç»“æŸæ—¶ Done)
    done = False
    
    # 8. è·å– S_{t+1}
    obs = self._next_observation()

    return obs, reward, done, {
        'gross_profit': gross_profit,
        'cost': cost,
        'price': price_curr, # è®°å½•å½“å‰çš„ç»“ç®—ä»·ç”¨äºç»˜å›¾ç­‰
        'scaled_pos': current_scaled_position
    }
```

```python
import numpy as np
import pandas as pd

class TradingMetrics:
    def __init__(self, returns, annualization=252):
        """
        åˆå§‹åŒ–è¯„ä»·æŒ‡æ ‡è®¡ç®—å™¨
        
        å‚æ•°:
            returns (pd.Series or np.array): æ—¶é—´åºåˆ—çš„å¥–åŠ±/å›æŠ¥åºåˆ— (R_t)
            annualization (int): å¹´åŒ–å› å­ï¼Œæ—¥çº¿æ•°æ®é€šå¸¸ä¸º 252
        """
        if isinstance(returns, list):
            returns = pd.Series(returns)
        self.returns = pd.Series(returns)
        self.freq = annualization
        
    def get_metrics_summary(self):
        """è¿”å›åŒ…å«æ‰€æœ‰è®ºæ–‡æŒ‡æ ‡çš„å­—å…¸"""
        return {
            "E(R)": self.expected_return(),
            "Std(R)": self.standard_deviation(),
            "DD": self.downside_deviation(),
            "Sharpe": self.sharpe_ratio(),
            "Sortino": self.sortino_ratio(),
            "MDD": self.max_drawdown(),
            "Calmar": self.calmar_ratio(),
            "% +ve Returns": self.win_rate(),
            "Ave. P / Ave. L": self.profit_loss_ratio()
        }

    def expected_return(self):
        """1. E(R): å¹´åŒ–æœŸæœ›å›æŠ¥"""
        # è®ºæ–‡å…¬å¼: mean(R) * 252
        return self.returns.mean() * self.freq

    def standard_deviation(self):
        """2. Std(R): å¹´åŒ–æ ‡å‡†å·®"""
        # è®ºæ–‡å…¬å¼: std(R) * sqrt(252)
        return self.returns.std() * np.sqrt(self.freq)

    def downside_deviation(self):
        """3. DD: ä¸‹è¡Œåå·® (Downside Deviation)"""
        # è®ºæ–‡å®šä¹‰: "annualised standard deviation of trade returns that are negative"
        # æå–è´Ÿæ”¶ç›Šéƒ¨åˆ†
        negative_returns = self.returns[self.returns < 0]
        
        if len(negative_returns) == 0:
            return 0.0
            
        # è®¡ç®—è´Ÿæ”¶ç›Šçš„æ ‡å‡†å·®å¹¶å¹´åŒ–
        # æ³¨æ„ï¼šè¿™é‡Œä¸¥æ ¼éµå¾ªè®ºæ–‡æ–‡å­—æè¿° "std of negative returns"
        # æ ‡å‡†é‡‘èå®šä¹‰é€šå¸¸æ˜¯ sqrt(mean(min(0, r)^2))ï¼Œä½†ä¸ºäº†å¤ç°è®ºæ–‡ï¼Œæˆ‘ä»¬è®¡ç®—è´Ÿæ ·æœ¬çš„std
        return negative_returns.std() * np.sqrt(self.freq)

    def sharpe_ratio(self):
        """4. Sharpe: å¹´åŒ–å¤æ™®æ¯”ç‡"""
        # è®ºæ–‡å…¬å¼: E(R) / Std(R)
        er = self.expected_return()
        std = self.standard_deviation()
        
        if std == 0: 
            return 0.0
        return er / std

    def sortino_ratio(self):
        """5. Sortino: ç´¢æè¯ºæ¯”ç‡"""
        # è®ºæ–‡å…¬å¼: E(R) / Downside Deviation
        er = self.expected_return()
        dd = self.downside_deviation()
        
        if dd == 0: 
            return 0.0
        return er / dd

    def max_drawdown(self):
        """6. MDD: æœ€å¤§å›æ’¤"""
        # å› ä¸ºè®ºæ–‡ä½¿ç”¨çš„æ˜¯ Additive Profits (åŠ æ³•åˆ©æ¶¦)ï¼Œå›æ’¤æ˜¯ç»å¯¹æ•°å€¼çš„ä¸‹é™
        # è®¡ç®—ç´¯è®¡è´¢å¯Œæ›²çº¿
        cumulative_wealth = self.returns.cumsum()
        
        # è®¡ç®—æˆªæ­¢ç›®å‰çš„æœ€é«˜ç‚¹ (High Water Mark)
        peak = cumulative_wealth.cummax()
        
        # è®¡ç®—å½“å‰å€¼ä¸æœ€é«˜ç‚¹çš„å·®å€¼
        drawdown = peak - cumulative_wealth
        
        # è·å–æœ€å¤§å›æ’¤å€¼
        mdd = drawdown.max()
        return mdd

    def calmar_ratio(self):
        """7. Calmar: å¡å°”ç›æ¯”ç‡"""
        # è®ºæ–‡å…¬å¼: Expected Annual Return / MDD
        er = self.expected_return()
        mdd = self.max_drawdown()
        
        if mdd == 0: 
            return 0.0
        return er / mdd

    def win_rate(self):
        """8. % +ve Returns: èƒœç‡"""
        # è®ºæ–‡å…¬å¼: æ­£æ”¶ç›Šæ¬¡æ•° / æ€»æ¬¡æ•°
        positive_count = (self.returns > 0).sum()
        total_count = len(self.returns)
        
        if total_count == 0: 
            return 0.0
        return positive_count / total_count

    def profit_loss_ratio(self):
        """9. Ave. P / Ave. L: ç›ˆäºæ¯”"""
        # è®ºæ–‡å…¬å¼: å¹³å‡æ­£æ”¶ç›Š / abs(å¹³å‡è´Ÿæ”¶ç›Š)
        positive_returns = self.returns[self.returns > 0]
        negative_returns = self.returns[self.returns < 0]
        
        if len(positive_returns) == 0:
            avg_positive = 0.0
        else:
            avg_positive = positive_returns.mean()
            
        if len(negative_returns) == 0:
            avg_negative = 0.0
        else:
            avg_negative = negative_returns.mean()
        
        # å¤„ç†è¾¹ç•Œæƒ…å†µ
        if avg_positive == 0.0 or avg_negative == 0.0: 
            return 0.0
        
        return avg_positive / abs(avg_negative)

# ==========================================
# æµ‹è¯•ä»£ç  (æ¨¡æ‹Ÿè®ºæ–‡ Table 2 çš„æ•°æ®æ ¼å¼)
# ==========================================
if __name__ == "__main__":
    # 1. åˆ›å»ºæ¨¡æ‹Ÿæ•°æ®
    # å‡è®¾æˆ‘ä»¬æœ‰ 1000 å¤©çš„äº¤æ˜“æ•°æ®ï¼Œå¹³å‡æ¯å¤©èµš 0.5ï¼Œæ³¢åŠ¨ç‡ä¸º 2.0
    np.random.seed(42)
    daily_rewards = np.random.normal(loc=0.05, scale=2.0, size=1000)
    
    # è½¬æ¢ä¸º Series
    rewards_series = pd.Series(daily_rewards)
    
    # 2. è®¡ç®—æŒ‡æ ‡
    metrics = TradingMetrics(rewards_series, annualization=252)
    results = metrics.get_metrics_summary()
    
    # 3. æ‰“å°ç»“æœ
    print("-" * 30)
    print("Paper Replication Metrics:")
    print("-" * 30)
    df_res = pd.DataFrame(results, index=['Value']).T
    print(df_res)
    
    print("-" * 30)
    print("Interpretation:")
    print(f"E(R) (å¹´åŒ–æ”¶ç›Š): {results['E(R)']:.4f}")
    print(f"Sharpe (å¤æ™®æ¯”ç‡): {results['Sharpe']:.4f}")
    print(f"MDD (æœ€å¤§å›æ’¤-ç»å¯¹å€¼): {results['MDD']:.4f}")
    
```
